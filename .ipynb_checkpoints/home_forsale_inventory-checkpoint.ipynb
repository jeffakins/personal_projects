## Zillow Pending Sales List

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
home_sales = pd.read_csv('for_sale_inventory_zillow.csv')
home_sales.head()
```

```python
home_sales.info()
```

```python
home_sales.shape
```

```python
home_sales.isnull().sum()
```

```python
home_sales = home_sales.dropna()
```

```python
home_sales.info()
```

```python
home_sales.head()
```

```python
city = home_sales.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'StateName'])
```

```python
city = city.rename(columns={'RegionName': 'city'})
```

```python
city = city.set_index('city')
```

```python
city.head()
```

```python
city = city.T
```

```python
city = city.rename_axis('date')
```

```python
city.head()
```

```python
city['Phoenix, AZ']
```

```python
#Plot 
plt.figure(figsize=(16, 9))
sns.set_theme(style="whitegrid")

# Plot the distribution of home inventory levels
sns.lineplot(data=city,
    palette="ch:rot=-.25,hue=1,light=.75",)

plt.title('Zillow Home Inventory Levels', fontsize = 20)
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Inventory', fontsize = 18)
plt.xticks(rotation = 45)
#plt.yticks()
#plt.grid()
plt.show()
```

```python
#Plot 
plt.figure(figsize=(16, 9))
sns.set_theme(style="whitegrid")

# Plot the distribution of home inventory levels
sns.lineplot(data=city[['Los Angeles-Long Beach-Anaheim, CA', 'San Diego, CA', 'San Jose, CA',
                      'Phoenix, AZ', 'Denver, CO', 'Salt Lake City, UT', 'Las Vegas, NV', 'Seattle, WA']],
             palette="tab10")

plt.title('Zillow Home Inventory Levels', fontsize = 20)
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Inventory', fontsize = 18)
plt.xticks(rotation = 45)
#plt.yticks()
#plt.grid()
plt.show()
```
